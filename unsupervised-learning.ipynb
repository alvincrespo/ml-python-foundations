# Unsupervised Learning

## What is unsupervised learning?

- Unsupervised learning is the process of building a descriptive model

> A descriptive model refers to a form of data analysis that aims to describe and make sense of large quantities of data, typically using techniques such as clustering or association rules.
>
> [ScienceDirect - Descriptive Model](https://www.sciencedirect.com/topics/computer-science/descriptive-model)

## Example

### Sample Dataset

| Customer | Spending | Income | Age | Gender | Education |
| --- | --- | --- | --- | --- | --- | --- |
| James S. | 7 | $40,000 | 22 | M | High School |
| Carol J. | 10 | $50,000 | 40 | F | Bachelors |
| Sarah P. | 2 | $36,000 | 53 | F | Bachelors |
| John J. | 3 | $72,000 | 24 | M | Doctorate |
| Saul B. | 1 | $18,000 | 18 | M | High School |
| Alan Q. | 9 | $98,000 | 32 | M | Doctorate |
| Frank D. | 8 | $28,000 | 19 | M | High School |
| Jane R. | 8 | $80,000 | 47 | F | Bachelors |
| Bill T. | 4 | $16,000 | 25 | M | High School |
| Tom C. | 5 | $63,000 | 62 | M | Masters |

Create a dictionary for the customers.

```python
data = [
  { "name": "James S.", "spending": 7, "income": 40000, "age": 22, "gender": "M", "education": "High School" },
  { "name": "Carol J.", "spending": 10, "income": 50000, "age": 40, "gender": "F", "education": "Bachelors" },
  { "name": "Sarah P.", "spending": 2, "income": 36000, "age": 53, "gender": "F", "education": "Bachelors" },
  { "name": "John J.", "spending": 3, "income": 72000, "age": 24, "gender": "M", "education": "Doctorate" },
  { "name": "Saul B.", "spending": 1, "income": 18000, "age": 18, "gender": "M", "education": "High School" },
  { "name": "Alan Q.", "spending": 9, "income": 98000, "age": 32, "gender": "M", "education": "Doctorate" },
  { "name": "Frank D.", "spending": 8, "income": 28000, "age": 19, "gender": "M", "education": "High School" },
  { "name": "Jane R.", "spending": 8, "income": 80000, "age": 47, "gender": "F", "education": "Bachelors" },
  { "name": "Bill T.", "spending": 4, "income": 16000, "age": 25, "gender": "M", "education": "High School" },
  { "name": "Tom C.", "spending": 5, "income": 63000, "age": 62, "gender": "M", "education": "Masters" }
]
```

### Plot the Dataset

```python
import matplotlib.pyplot as plt

# Extract income and spending from data
incomes = [customer['income'] for customer in data]
spendings = [customer['spending'] for customer in data]

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(incomes, spendings)

# Add customer names under each point
for customer in data:
  plt.text(customer['income'], customer['spending'], customer['name'],
       ha='center', va='top', fontsize=8)

plt.xlabel('Income ($)')
plt.ylabel('Spending')
plt.title('Customer Income vs Spending')
plt.grid(True, alpha=0.3)
plt.show()
```

### Segment the Customers

```python
import matplotlib.pyplot as plt

# Define segments
# These segments are hypothetical and for illustration purposes
# In a real scenario, segments would be determined by a clustering algorithm by the unsupervised learning process
segments = {
  'Segment 1': ['Frank D.', 'James S.', 'Carol J.'],
  'Segment 2': ['Bill T.', 'Saul B.', 'Sarah P.'],
  'Segment 3': ['John J.', 'Tom C.', 'Jane R.', 'Alan Q.']
}

# Create scatter plot
plt.figure(figsize=(10, 6))

# Plot each segment with different colors
colors = ['red', 'blue', 'green']
for idx, (segment_name, customer_names) in enumerate(segments.items()):
  segment_incomes = [c['income'] for c in data if c['name'] in customer_names]
  segment_spendings = [c['spending'] for c in data if c['name'] in customer_names]
  plt.scatter(segment_incomes, segment_spendings, s=200, alpha=0.3,
        color=colors[idx], label=segment_name, edgecolors=colors[idx], linewidths=2)

# Plot all points
plt.scatter(incomes, spendings, s=50, color='black', zorder=5)

# Add customer names under each point
for customer in data:
  plt.text(customer['income'], customer['spending'], customer['name'],
       ha='center', va='top', fontsize=8)

plt.xlabel('Income ($)')
plt.ylabel('Spending')
plt.title('Customer Income vs Spending - Segmented')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
